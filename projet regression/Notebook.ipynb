```python
#Stating the problem

y = data["GPA"].values.reshape(-1,1) #matrice colonne 
print(y)
x= data.drop(columns = ["GPA", "StudentID"]).values

#ajouter la constante (modèle avec constante)
ones = np.ones((x.shape[0], 1))
x = np.hstack((ones,x))
print(x)
X_df = pd.DataFrame(x, columns = ['const'] + data.drop(columns=["GPA", "StudentID"]).columns.tolist())
X_df
```

```python
def regression(y,x):
    A = np.linalg.inv(x.T @ x) @ x.T @ y
    y_chapeau = x @ A
    return y_chapeau
```

```python
def SCR(y,x):
    return np.linalg.norm(y - regression(y,x))**2

def SCE(y,x):
    return np.linalg.norm(regression(y,x) - y.mean())**2
def SCT(y,x):
    return np.linalg.norm(y - y.mean())**2
```

```python
def R(y,x): #coef de determination
    return 1 - SCR(y,x)/SCT(y,x)

def R_ajuste(y,x):
    return 1- ((len(y)-1)/(len(y)-x.shape[1]-1)) * (1 - R(y,x))
```

```python
R_ajuste(y,x)
```

```python
def etape1(y, x):
    y_centre = y - y.mean()
    corr = {}

    for i in x.columns:
        if i == "const":
            continue  # ne pas inclure l’interception

        x_i = x[i].values.reshape(-1, 1)
        x_i_centre = x_i - x_i.mean()

        num = (y_centre * x_i_centre).sum()  # ✔️ produit terme à terme
        denom = np.sqrt((y_centre**2).sum()) * np.sqrt((x_i_centre**2).sum())


        rho = num / denom
        corr[i] = abs(rho)

    return max(corr, key=corr.get)

```

```python
etape1(y,X_df)
```

```python
def corr_partielle(ancien,nouveau):
    return (SCR(y,ancien) - SCR(y,nouveau))/SCR(y,ancien)

def fisher(ancien, nouveau, ddl):
    return (SCR(y,ancien) - SCR(y,nouveau))/(SCR(y,nouveau)/ddl)
```

```python
from scipy.stats import f
def methode_progressive(y, x, alpha=0.05):
    n = len(y)
    selected = []
    variables = list(x.columns[1:])  # on saute "const"
    # Étape 1 : choisir la variable la plus corrélée
    first_var = etape1(y, x)
    selected.append(first_var)
    variables.remove(first_var)
    print(f"✅ 1ère variable sélectionnée : {first_var}")
    while True:
        best_r2 = -np.inf
        best_var = None
        x_ancien = x[selected]
        for var in variables:
            x_nouveau = x[selected + [var]]
            r2 = corr_partielle(x_ancien.values, x_nouveau.values)
            if r2 > best_r2:
                best_r2 = r2
                best_var = var
        # Tester la meilleure variable trouvée
        x_nouveau = x[selected + [best_var]]
        ddl = n - x_nouveau.shape[1] - 1
        F_obs = fisher(x_ancien.values, x_nouveau.values, ddl)
        F_crit = f.ppf(1 - alpha, 1, ddl)
        print(f" Candidat : {best_var} | F = {F_obs:.3f} | F_crit = {F_crit:.3f}")
        if F_obs > F_crit:
            print(f"Variable retenue : {best_var}")
            selected.append(best_var)
            variables.remove(best_var)
        else:
            print("Aucune variable significative. Fin de la sélection.")
            break
    return selected


```

```python
new_columns = methode_progressive(y, X_df)
new_columns
```

```python
x_nouveau = X_df[new_columns]
x_nouveau = np.hstack((np.ones((x_nouveau.shape[0], 1)), x_nouveau))
x_nouveau
```
